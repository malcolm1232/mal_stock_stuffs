# Add PE Ratio to DataFrame

```python
pe_ratio_list = [] #trailingPE #marketCap
able_list = []
unable_list = []
for ticker in df['colName']:
    print(ticker)
    try:
        df_temp_info = pd.read_csv('../obtain_data/ticker_data_info/ticker_data_Daily_2021-12-12/{}_ended_2021-12-12_info.csv'.format(ticker))
        pe_ratio = df_temp_info['trailingPE'][0]
        print('ABLE Ticker: {}'.format(ticker))
        pe_ratio_list.append(pe_ratio)
        able_list.append(ticker)
    except:
        print('unable Ticker: {}'.format(ticker))
        pe_ratio_list.append(ticker)
        unable_list.append(ticker)
pe_ratio_list
```

# Add Market Cap to DataFrame

```python
marketCap_list = [] #trailingPE #marketCap
able_list = []
unable_list = []
for ticker in df['colName']:
    print(ticker)
    try:
        df_temp_info = pd.read_csv('../obtain_data/ticker_data_info/ticker_data_Daily_2021-12-12/{}_ended_2021-12-12_info.csv'.format(ticker))
        pe_ratio = df_temp_info['marketCap'][0]
        print('ABLE Ticker: {}'.format(ticker))
        marketCap_list.append(pe_ratio)
        able_list.append(ticker)
    except:
        print('unable Ticker: {}'.format(ticker))
        marketCap_list.append(0)
        unable_list.append(ticker)
marketCap_list
```

# Add Industry to Track PE 
- well.., every industry has different average PEs

```python
industry_list = [] #trailingPE #marketCap
able_list = []
unable_list = []
for ticker in df['colName']:
    print(ticker)
    try:
        df_temp_info = pd.read_csv('../obtain_data/ticker_data_info/ticker_data_Daily_2021-12-12/{}_ended_2021-12-12_info.csv'.format(ticker))
        pe_ratio = df_temp_info['industry'][0]
        print('ABLE Ticker: {}'.format(ticker))
        industry_list.append(pe_ratio)
        able_list.append(ticker)
    except:
        print('unable Ticker: {}'.format(ticker))
        industry_list.append(0)
        unable_list.append(ticker)
industry_list
```

```python
dfpe_ratio_list = pd.DataFrame(pe_ratio_list,columns = ['PE_RATIO'])
dfmarketCap_list = pd.DataFrame(marketCap_list,columns = ['marketCap'])
dfindustry_list = pd.DataFrame(industry_list,columns = ['industry'])
```

```python
dfcp1_pe = pd.concat([dfcp1, dfpe_ratio_list,dfmarketCap_list,dfindustry_list],axis=1)
dfcp1_pe
```

```python
dfcp1_pe.fillna(0, inplace=True)
```

```python
dfcp1_pe['marketCap'] = dfcp1_pe['marketCap'].apply(lambda x: int(x)/1000000)
dfcp1_pe
```

# How to Analyze

#### Column Names:
#### Price Started Crashing on ~20 Feb 2020 
#### Price Recovery Date: ~27 March 2020
- colName : Ticker Symbol
- ticker_min_list: Bottom Price in ~March 2020 (Gotten from min from Closing Price), so not technically a "Low"
- ticker_max_list: Max Price in February 2020 (Gotten from max from Closing Price), so not technically a "High"
- Low_from_High_in_percentage: At the bottom of the crash; i.e ticker_min_list, Percentage of Initial Value on 20 March 2020 ; 0.4 == on 27 March 2020, it was 40% of the value on 20th Feb 2020
- value_lost_in_percentage: Amount of MarketCap Shaved off; 0.4 = 40% Lost
- current_price: Current Price as of Data of Scraping. (Time of Writing: 13 Dec 2021)
- current_P_from_PostCrashInMarch2020_%: Price Increment from 27 March 2020 to Current Price; 0.4 = 40%
- current_P_from_PreCrashIn20Feb2020_%: Price Increment from 20 Feb 2020 to Current Price i.e Ceteris Paribus and price did not crash. stock was en-route doing its own Thang; 0.4 = 40%
    - Hypothesis: To compare Economic Recovery of pre and post covid with Stock as reference    
- PE_RATIO: PE ratio
- marketCap: self-explanatory
- S&P_500: Boolean, True/False If Ticker is in S&P_500
