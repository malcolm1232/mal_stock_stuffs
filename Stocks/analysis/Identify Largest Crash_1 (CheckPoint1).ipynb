```python
for count in range(len(ticker_csv_list)):#[:50]
    
    if count == 0:
        print("count:",count)
        csv_path = os.listdir(datapath)[count] #change
#         print("csv_path:",csv_path)
        ticker = csv_path.split('_ended')[0]
#         print("ticker:",ticker)
        df_current = pd.read_csv(datapath + csv_path)
        df_current_DC = df_current[['Date','Close']]
        df_current_DC.columns = ['Date', ticker]
        df_current_DC
        
        
        join_df = df_date.merge(
            df_current_DC,
            how="left",
            left_on="Date", # col name on df1; df
            right_on="Date", # col name on df2;df_updated
        )
        df = join_df
#         print("df:",df)
    else:
        
        print("count:",count)
        csv_path = os.listdir(datapath)[count] #change
#         print("csv_path:",csv_path)
        ticker = csv_path.split('_ended')[0]
#         print("ticker:",ticker)
        df_current = pd.read_csv(datapath + csv_path)

        df_current_DC = df_current[['Date','Close']]
        df_current_DC.columns = ['Date', ticker]
        df_current_DC
        
        join_df = df_current_DC.merge(
            df,
            how="right",
            left_on="Date", # col name on df1; df
            right_on="Date", # col name on df2;df_updated
        )
#         print("join_df:\n",join_df)
        df = join_df
        
```

```python
df_short = df.copy()
```

# Stock Crash of 20 February 2020

```python
# df_short['Date'] == ['2020-02-20']
df_crashdate = df_short[df_short['Date']== '2020-02-20'].index[0]
df_recoverydate = df_short[df_short['Date']== '2020-03-27'].index[0]
print("df_crashdate:",df_crashdate)
print("df_recoverydate:",df_recoverydate)
```

```python
df_short = df_short[df_crashdate:df_recoverydate]
df_short
```

# Section 1. Get Min , Get Max. Calculate largest price drop in %
# Section 2. Get Largest Increase from the min to current date

# Identify Largest PE Ratio Drop


```python
df_ = df_short.copy()
colName = df_.columns.to_list()[1:]

ticker_min_list = []
ticker_max_list = []
print("colName:",colName)
for ticker in colName:
    print("ticker:",ticker)
    ticker_min = min(df_[ticker])
    ticker_max = max(df_[ticker])

    ticker_min_list.append(ticker_min)
    ticker_max_list.append(ticker_max)
```

```python
print("ticker_min_list:",ticker_min_list)
print("ticker_max_list:",ticker_max_list)
```

# Calculation

```python
Low_from_High_in_percentage = [min_ / max_ for min_, max_ in zip(ticker_min_list, ticker_max_list) ] #get Low (%) from ATH 
print("Low_from_High_in_percentage:",Low_from_High_in_percentage)
value_lost_in_percentage = [1 - low_ for low_ in Low_from_High_in_percentage] # Value Lost in % i.e 1- (get Low (%) from ATH)
print("value_lost_in_percentage:",value_lost_in_percentage)
```

```python
Low_from_High_in_percentage_df = pd.DataFrame(Low_from_High_in_percentage, columns = ['Low_from_High_in_percentage'])
value_lost_in_percentage_df = pd.DataFrame(value_lost_in_percentage, columns = ['value_lost_in_percentage'])
ticker_min_list_df = pd.DataFrame(ticker_min_list, columns = ['ticker_min_list'])
ticker_max_list_df = pd.DataFrame(ticker_max_list, columns = ['ticker_max_list'])
df_colName = pd.DataFrame(colName, columns = ['colName'])

df_concat = pd.concat([df_colName,ticker_min_list_df,ticker_max_list_df,Low_from_High_in_percentage_df,value_lost_in_percentage_df], axis=1)
df_concat
```

```python
df_checkpoint1 = df_concat.copy()
dfcp = df_checkpoint1
dfcp
```

```python
dfcp.dropna(axis=0, inplace = True)
dfcp
```

```python
dfcp.sort_values(by = 'value_lost_in_percentage',ascending = False)
```

```python
dfcp.head()
dfcp['colName'].to_list()
```

```python
# Get Current Price
curren_price_list = []
for dfcp_ticker in dfcp['colName'].to_list():
    print(dfcp_ticker)
#     dfcp_ticker = 'AAPL'
    temp_df = pd.read_csv('../obtain_data/ticker_data/ticker_data_Daily_2021-12-11/{}_ended_2021-12-11.csv'.format(dfcp_ticker))
    temp_df
    closing_price = temp_df.loc[len(temp_df)-1,'Close'] # Get Current Closing Price
    curren_price_list.append(closing_price)
```

```python
df_curren_price_list= pd.DataFrame(curren_price_list, columns = ['current_price'])
df_curren_price_list
dfcp.reset_index(drop = True, inplace=True)
dfcp
dfcp1 = pd.concat([dfcp,df_curren_price_list ],axis=1)
dfcp1
```

```python
# dfcp1 = dfcp1.iloc[:, :-1]
```

```python
# Calculation: We are trying to calculate how much the price has risen from the MINIMUM; the bottom of the crash in 27 March 2020
dfcp1['current_P_from_PostCrashInMarch2020_%'] = list(list(dfcp1['current_price'] - dfcp1['ticker_min_list'])/ dfcp1['ticker_min_list'])
dfcp1
```

```python
# Calculation: We are trying to calculate how much the price has risen from the MAXIMUM; Pre-Crash of March 2020. 
dfcp1['current_P_from_PreCrashIn20Feb2020_%'] = list(list(dfcp1['current_price'] - dfcp1['ticker_max_list'])/ dfcp1['ticker_max_list'])
dfcp1 #I.E did it return to current levels?
```
