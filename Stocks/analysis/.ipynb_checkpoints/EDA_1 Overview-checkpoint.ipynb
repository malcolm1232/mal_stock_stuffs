```python
csvfname ='Identify Largest Crash_1 (CheckPoint2)_SNP500.csv'
df = pd.read_csv(csvfname,index_col = 0)
df.head(2)
```

```python
df.drop(['Low_from_High_in_percentage', 'S&P_500'],axis=1,inplace=True)
df.head(2)
```

```python
colName=['tick', 'min', 'max', 'val_lost','Price', 'L_to_Price','H_to_Price', 'PE', 'MktCap','industry']
```

```python
df.columns = colName
df.reset_index(drop=True,inplace=True)
df
```

```python
tickerlist=df['tick'].to_list()
tickerlist
```

```python
from urllib.request import Request, urlopen

def get_soup(url):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    web_byte = urlopen(req).read()
    html_doc = web_byte.decode('utf-8')
    soup = bs.BeautifulSoup(html_doc, 'html.parser')
    return soup
import requests
import bs4 as bs
import pandas as pd
# tickerlist = ['AAPL','V']
PE_list = []
able_list = []
unable_list = []
for count, ticker in enumerate(tickerlist):#shorten
    count+=1
    try:
        url = 'https://finance.yahoo.com/quote/{}/'.format(ticker)
        soup=get_soup(url)
        for i in soup.findAll('td',{'data-test':"PE_RATIO-value"}):
            PE = i.text
#             print(PE)
        PE_list.append(PE)
        able_list.append(PE)
        print('{} / {} done, [{}] PE: {}'.format(count, len(tickerlist), ticker, PE))
    except:
        unable_list.append(PE)
        able_list.append(ticker)
        print('{} / {} Unable, [{}] PE: {}'.format(count, len(tickerlist), ticker, PE))
        pass
df_PE_list = pd.DataFrame(PE_list, columns=['PE'])
df_PE_list
```

# Plot Stocks that have more than doubled in price from 20 Feb 2020
#### I.e, Hypothetically, company is "healthier" during pandemic.

```python
stocks_that_doubled = df_HtoPrice_sorted[df_HtoPrice_sorted['H_to_Price'] >= 1.0]['tick'].to_list()
df_HtoPrice_sorted[df_HtoPrice_sorted['H_to_Price'] >= 1.0]
```

```python
def plot_stocks(ticker):
    
    temp_industry = list(df[df['tick'] == ticker]['industry'])[0]
    
    tempdf = pd.read_csv('../obtain_data/ticker_data/ticker_data_Daily_2021-12-11/{}_ended_2021-12-11.csv'.format(ticker))
    title_ = '{}, {}'.format(ticker,temp_industry)
    plt.title(title_)
    index_20Feb = list(tempdf[tempdf['Date'] == '2020-02-20'].index)[0]
    plt.axvline(x=index_20Feb, ls='--',linewidth=0.5, color='r' )
    plt.plot(tempdf['Close'])
    plt.grid(linestyle='--', linewidth=0.5)
    plt.show()
for ticker in stocks_that_doubled:
    print("ticker:",ticker)
    plot_stocks(ticker)
```
