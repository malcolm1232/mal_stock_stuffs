```python
import datetime as dt
today_date = dt.datetime.now().strftime('%Y-%m-%d')
import os
dirpath_to_save = './ticker_data_info/ticker_data_Daily_{}/'.format(today_date)
os.makedirs(dirpath_to_save)
```
