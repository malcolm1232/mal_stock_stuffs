```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.6f' % x)
```

```python
pathtosave = './ticker_data/ticker_data_2021-09-25_processed/all_stock_price_info__stocks_at_ATL.csv'
mainpathtosave = './ticker_data/ticker_data_2021-09-25_processed/'
df= pd.read_csv('./ticker_data/ticker_data_2021-09-25_processed/all_stock_price_info.csv')
```

```python
#get new column, get perc_from_ATH_ < 5%
df['perc_from_ATH_'] = df['perc_from_ATH']*100
print(df)
df1 = df[df['perc_from_ATH_'] <5]
```

```python
#relocate columns only
df_atl = pd.DataFrame(df1['perc_from_ATH_'])
df1.drop('perc_from_ATH_', axis=1, inplace=True)
df2 = pd.concat([df_atl, df1], axis=1)
df2.drop('Unnamed: 0', axis=1, inplace=True)
df = df2.copy()
```

```python
df_tempo = df['start_date'].apply(lambda x: x.split('-')[0]).copy()
df.drop('start_date', axis=1, inplace = True)
df = pd.concat([df_tempo, df], axis=1)
df
```

```python
df.to_csv(pathtosave)
```

```python
#Want only companies that have existed since 1999
df['start_date'] = df['start_date'].apply(lambda x: int(x))
df1 = df[df['perc_from_ATH_'] > 1.5]# want at least 1.5% founded since 2000 or less
```

```python
df1 = df1[df1['start_date'] <=2000] # want co. founded since 2000 or less
df1.to_csv(mainpathtosave + 'stocks_at_ATL_since2000_MoreThan1.5percFromATL.csv')
print(df1)
```
