```python
#get 20%, 40%, 60%, 80%, 100% markup from minimum

def get_entire_df(df, ticker):
    maxall = df['High'].max()
    minall = df['Low'].min()
#     df_csv = pd.read_csv('C:/Users/malco/Desktop/Server Data 2/all stock price/a.csv')

    threshold = list(np.arange(.1,1.1,.1))
#     print('threshold:',threshold)
    diff = maxall - minall

    em = [] #to contain all threshold values; markup from min to max by 10%
    em.append(ticker)
    
    print('tolistdate:',df['Date'])
    em.append(df['Date'].to_list()[0])
    em.append(df['Adj Close'].to_list()[-1])
    em.append(minall)
    em.append(maxall)
    em.append(diff*.1)

    for threshold_ in threshold:
        value = threshold_ * diff
#         print('value:', value)
        value += minall
        em.append(value)
    df_em = pd.DataFrame(em).T

    df_basic = pd.DataFrame()
    df = pd.concat([df_basic, df_em], axis=0)



    df.columns = ['stock','start_date','current_price','minall','maxall','each_markup','markup_10','markup_20','markup_30','markup_40',
                    'markup_50','markup_60','markup_70','markup_80','markup_90','markup_100']
    print(df)
    # df.index = ['a']
    
    return df


```

```python
#create max column
for i in os.listdir(path)[1:]:#shorten
    
    ticker = i.split('.')[0]
    df_csv = pd.read_csv(path+i)
    print(path+i)
    print(df_csv)
    df_each = get_entire_df(df_csv,ticker)
    print(df_each)
    df1 = pd.concat([df1, df_each],axis=0, sort = False)
    
print(df1)
    
    
```
