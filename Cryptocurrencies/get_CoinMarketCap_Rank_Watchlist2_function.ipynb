# Get Cryptocurrency Number Watchlist from CoinMarketCap

```python
import bs4 as bs
import requests
```

```python
url = 'https://coinmarketcap.com/?page={}'
# resp = requests.get(url)
# soup = bs.BeautifulSoup(resp.text, 'lxml')
```

```python
# get /currencies/btc 
count = 0
cryp_list = []
```

```python
for page_num in range(1, 50+1):#shorten
    count = 0
    print('page_num:', page_num)
    url = 'https://coinmarketcap.com/?page={}'.format(page_num)
    resp = requests.get(url)
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    print(url)
    for i in soup.findAll('a', class_ = "cmc-link"):
        href_primary = i.get('href')
        href_sec = href_primary.split('/')[1]
        if href_sec == "currencies":
    #         print('count: {} , {}' .format(count,href_primary))
            if len(href_primary.split('/')) == 4:
                print('page: {} count: {} , {}' .format(page_num, count,href_primary))
                cryptocurrency_ticker = href_primary.split('/')[2]
                if cryptocurrency_ticker not in cryp_list:
                    cryp_list.append(cryptocurrency_ticker) 
                    print('appending: ',cryptocurrency_ticker)
        count+=1
    print('\n_____NEXT PAGE_____')
```

```python
cryp_list
```

```python
# get temp_url
mainurl = 'https://coinmarketcap.com/'
mainurl_1 = 'currencies'

rank_list = []
watchlist_list = []
marketcap_list = []

for i in range(len(cryp_list)):#shorten
    temp_cryp = cryp_list[i]
    temp_url = mainurl + mainurl_1 + '/' + temp_cryp
    temp_url
    
    temp_resp = requests.get(temp_url)
    temp_soup = bs.BeautifulSoup(temp_resp.text, 'lxml')
    temp_soup

    for i in temp_soup.findAll('div', class_= 'namePill'):
        if 'Rank #' in i.text:
            print(i.text)
            rank_list.append(i.text)
        if 'watchlists' in i.text:
            watchlist_list.append(i.text)    
    
    # for i in soup.findAll('div', class_ = 'statsValue')[0]:
#         marketcap = i.text
#         marketcap_list.append(marketcap)
    #     marketcap = int(i.text.replace(',','').replace('$', ''))
    #     print(marketcap)
```

```python
import pandas as pd
dfcryp_list = pd.DataFrame(cryp_list, columns = ['cryp_list'])
dfrank_list = pd.DataFrame(rank_list, columns = ['rank_list']) 
dfwatchlist_list = pd.DataFrame(watchlist_list, columns = ['watchlist_list'])
df_concat = pd.concat([dfcryp_list,dfrank_list, dfwatchlist_list],axis=1)
df_concat = df_concat[:-1]
df_concat
```

```python
df_concat.to_csv('CMC_Ranking.csv')
```
